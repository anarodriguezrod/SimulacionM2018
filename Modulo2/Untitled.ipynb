$m\ddot{x} + kx + B\dot{x}= 0$

Requerimos la solucion de

$$\dot {x} = y$$

$$\dot {y} = -\frac{k}{m}x - \frac{B}{m}y$$ 

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from scipy.integrate import odeint
```

```python
def armonico (y,t,k,m,B):
    y1 = y[0]
    y2 = y[1]
    return y2, -k*y1/m - B/m * y2
```

#### Condiciones iniciales 

```python
y0 = [.6,.8]
k = 3.0
m = 1.0
B = .5
```

```python
tiempo = np.linspace(0, 50, 500)
```

```python
solucion = odeint(armonico, y0, tiempo, args=(k,m,B))
```

```python
solucion.shape
```

```python
posicion = solucion[:,0]
velocidad = solucion[:,1]
```

```python
plt.plot(tiempo, posicion, c = "r", label = "Posición")
plt.plot(tiempo, velocidad, c = "k", label = "Velocidad")
plt.legend()
plt.xlabel("tiempo")
plt.show()
```

```python
def amortiguado (t=0):
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.plot(tiempo, posicion, "k-")
    ax.plot(tiempo[t], posicion[t], "ro")
    ax.text(4, .6, "tiempo = %s" %tiempo[t])
    ax.text(4, .5, "posicion = %s" %posicion[t])  
    fig.canvas.draw()
    
```

$\omega_0^2 = k/m$,$\Gamma = \frac{B}{2m}$

1. subamortiguado, $\omega_0^2 > \Gamma^2$
2. sobreamortiguadp, $\omega_0^2 > \Gamma^2$
3. amortiguado crítico, $\omega_0^2 > \Gamma^2$

## Subamortiguado

```python
y0 = [.6,.4]
k = 0.3
m = 1.0
B = .5
tiempo1 = np.linspace(0, 50, 500)
solucion1 = odeint(armonico, y0, tiempo1, args=(k,m,B))
posicion1 = solucion1[:,0]
velocidad1 = solucion1[:,1]
plt.plot(tiempo, posicion1, c = "r", label = "Posición")
plt.plot(tiempo, velocidad1, c = "k", label = "Velocidad")
plt.legend()
plt.xlabel("tiempo")
plt.show()
```

Si el numero aleatorio es menor a .5 entonces el paso es hacia la izquierda, y si es mayor o igual a .5 sera hacia la derecha

```python
pasosderecha = 0
pasosizquierda = 0
pasos = 100
x0= 0
posiciones = [x0]
for paso in range(pasos):
    decision = np.random.rand()
    if decision < .5:
        pasosizquierda +=1
        x0-= 1
        posiciones.append(x0)
    else: 
        pasosderecha +=1
        x0 += 1
        posiciones.append(x0)
```

# Actividad

```python
def borracho(pasos):
    pasosderecha = 0
    pasosizquierda = 0
    x0= 0
    posiciones = [x0]
    for paso in range(pasos):
        decision = np.random.rand()
        if decision < .5:
            pasosizquierda +=1
            x0-= 1
            posiciones.append(x0)
        else: 
            pasosderecha +=1
            x0 += 1
            posiciones.append(x0)
               
    return posiciones
```

10 borachos!!

```python
for r in range(5):
    plt.plot(borracho(100), lw = .3, label= "Borracho %s" % r)
    plt.xlabel("Pasos")
    plt.ylabel("Posicion")
plt.legend()
plt.show()
```
